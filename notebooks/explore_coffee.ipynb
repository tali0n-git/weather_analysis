```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isna().sum()
```

```python
# Looking at null-containing rows only
coffee[coffee.isna().any(axis=1)]
```

```python
# TODO: drop your null values
clean_coffee = coffee.dropna()
clean_coffee.isna().sum()
```

```python
# TODO: Select only Minas Gerais

clean_coffee = clean_coffee[clean_coffee["subdivision"] == "Minas Gerais"]
clean_coffee.reset_index(drop=True, inplace=True)
clean_coffee
```

```python
# TODO: BEGIN YOUR EDA!

sns.lineplot(data=clean_coffee, x="year", y="bearing_trees")

sns.lmplot(data=clean_coffee, x="year", y="bearing_trees")

coeffs = np.polyfit(clean_coffee["year"].values, clean_coffee["bearing_trees"].values, 1)
slope = coeffs[0]
intercept = coeffs[1]
print(f"Equation of the line of best fit: y = {slope:.2f}x + {intercept:.2f}")
```

```python
clean_coffee.describe()

mean_trees = clean_coffee.describe().iloc[1, 3]
med_trees = clean_coffee.describe().iloc[5, 3]

# Is mean greater, less than or similar to the median?

print(f"Mean: {mean_trees}")
print(f"Median: {med_trees}")
print(f"Difference: {mean_trees - med_trees}")
```

```python
sns.histplot(clean_coffee["bearing_trees"], kde=True)
plt.xlabel("bearing_trees")
plt.ylabel("Frequency")
plt.show()
```
